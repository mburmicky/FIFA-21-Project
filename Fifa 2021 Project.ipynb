### `Cleaning the dataframe before building the prediction model`

- Cleaning all object features with '+' sign

```python
#First we decided to clean all object columns with '+' sign, and keep the value on the 'left'

cols_to_clean =['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','GK']

for column in cols_to_clean:
    fifa_dropped[column] = fifa_dropped[column].str.split('+').str[0]
```

```python
#Then we convert the same columns into numerical features
cols_to_clean =['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','GK']

for col in cols_to_clean:
    fifa_dropped[col] =  pd.to_numeric(fifa_dropped[col], errors='coerce')
```

```python
#Now we revise that the features type, were converted correctly
fifa_dropped.dtypes
```

```python
#Here we further analyze using 'value_counts' function the values contained in rest of features
for col in fifa_dropped:
    print(fifa_dropped[col].value_counts(dropna=False))
    print(     )
```

- Cleaning height

```python
#cleaning height. Converting it from feet-inches to cm

fifa_dropped['Height']=fifa_dropped['Height'].str.replace('"','').astype(str)
display(fifa_dropped['Height'])

def convert_to_cm(h):
    feet, inches = map(int,h.split("'"))
    tot_inch = feet*12 + inches
    return tot_inch * 2.54

fifa_dropped['Height']=fifa_dropped['Height'].apply(convert_to_cm)
display(fifa_dropped['Height'])
```

- Cleaning 'Value','Wage','Release Clause'

```python
#removing € sing, and fitting 'k', 'M', in same scale
value_columns = ['Value','Wage','Release Clause']
for col in value_columns:
    fifa_dropped[col] = fifa_dropped[col].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
fifa_dropped
```

- Cleaning 'weight'

```python
#From pounds to kilogram (so it is consistent with height conversion in 'cm')

fifa_dropped['Weight']=fifa_dropped['Weight'].str.replace('lbs','').astype(int)*0.453592
display(fifa_dropped['Weight'])
```

```python
#We re-visualize features types, to understand what needs further cleaning
fifa_dropped.dtypes
```

```python
#We re-visualize values of each features, to understand what needs further cleaning

for col in fifa_dropped:
    print(fifa_dropped[col].value_counts(dropna=False))
    print(     )
```

- Cleaning features with '★' sign

```python
fifa_dropped['SM']=fifa_dropped['SM'].str.replace('★','').astype(int)
fifa_dropped['SM']
```

```python
star_clean_col=['W/F','IR']
for col in star_clean_col:
    fifa_dropped[col]=fifa_dropped[col].str.replace(' ★','').astype(int)
```

```python
#Here we check the result through value counts
for col in fifa_dropped:
    print(fifa_dropped[col].value_counts(dropna=False))
    print(     )
```

### `Scaling numerical features and encoding categoricals`

```python
#Before X-y split, we scale numerical features (not including OVA as it is the target), using the normalized method

#First we drop OVA
numerical_but_OVA = numerical.drop(columns=['OVA'])

#Bringing libraries for scaling
from sklearn.preprocessing import MinMaxScaler 

#Now we scale it, using MinMax scaler method

transformer = MinMaxScaler().fit(numerical_but_OVA)
numerical_but_OVA_norm = transformer.transform(numerical_but_OVA)
print(numerical_but_OVA_norm.shape)

numerical_but_OVA_norm=pd.DataFrame(numerical_but_OVA_norm, columns=numerical_but_OVA.columns)

#and here we display it to check the result
numerical_but_OVA_norm
```

```python
#Here we encode all categorical features,into numerical features
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categorical)

cols = encoder.get_feature_names_out(input_features=categorical.columns) #this code line is key to get all the column names in the variable cols

categorical_encoded = pd.DataFrame(encoder.transform(categorical).toarray(),columns=cols)

#and here we display it to check the result
categorical_encoded
```

### `X-y split`

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

#Now lets X-y split

y=numerical['OVA']


#And to define X, we concatenate dataframes (numerical_but_OVA_norm and categorical_encoded)

X=pd.concat([numerical_but_OVA_norm, categorical_encoded], axis=1)

#and here we display X to check the concatenation result

X
```

### `Linear Regression - Train-test split`

```python
# Here we split the training and test data randomly from the data frame, to avoid having an unrepresentative prediction model

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#we now train/fit our second model: linear regression (just as a reminder, lm only considered numerical variables)
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

### `Linear Regression - Checking precision of the model`

```python
#Here we check the r2 score from the 'train' data to get an understanding of the precision

from sklearn.metrics import r2_score
predictions = lm.predict(X_train)

#Here we check r2 score with training data
r2_score(y_train, predictions)
```

```python
#Here we check r2 score with test data. Surprinsingly, the 'test' r2 score, is higuer that the one using training data. Ussually, is the oppossite 
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

### `Defining clean function`

```python
def clean (df):
    #Dropping features
    cols_to_drop = ['Hits','Name','Nationality','Club','Position','Team & Contract','Joined','Loan Date End','Contract','Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control','Acceleration','Sprint Speed','Agility','Reactions','Balance','Shot Power','Jumping','Stamina','Strength','Long Shots','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Marking','Standing Tackle','Sliding Tackle','GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes','PAC','SHO','PAS','DRI','DEF','PHY']
    df = df.drop(cols_to_drop, axis=1)
    
    #First we decided to clean all object columns with '+' sign, and keep the value on the 'left'    
    cols_to_clean =['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','GK']
    for column in cols_to_clean:
        df[column] = df[column].str.split('+').str[0]
    
    #Then we convert the same columns into numerical features
    cols_to_clean =['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','GK']
    for col in cols_to_clean:
        df[col] =  pd.to_numeric(df[col], errors='coerce')
    
    #cleaning height. Converting it from feet-inches to cm
    
    df['Height']=df['Height'].str.replace('"','').astype(str)
    
    def convert_to_cm(h):
        feet, inches = map(int,h.split("'"))
        tot_inch = feet*12 + inches
        return tot_inch * 2.54
    df['Height']=df['Height'].apply(convert_to_cm)
    
    #removing € sing, and fitting 'k', 'M', in same scale
    
    value_columns = ['Value','Wage','Release Clause']
    for col in value_columns:
        df[col] = df[col].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
    
    #From pounds to kilogram (so it is consistent with height conversion in 'cm')
    
    df['Weight']=df['Weight'].str.replace('lbs','').astype(int)*0.453592
    
    #Cleaning features with '★' sign
    df['SM']=df['SM'].str.replace('★','').astype(int)
    
    star_clean_col=['W/F','IR']
    for col in star_clean_col:
        df[col]=df[col].str.replace(' ★','').astype(int)
    
    #THIS STEP WORKED for my previous df. Should we adapt it to the new data frame?
    #Here we replace the Nan values of the features'A/W','D/W', with the most repetitive value of these columns ('Medium')
    
    nan_clean_col=['A/W','D/W']
    for col in nan_clean_col:
        df[col]=df[col].fillna("Medium")
    
    return df
```
